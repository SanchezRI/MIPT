#### Task 4.3

```python
third_row_sec_col = rand_arr[2][1]
crop_first_two_rows = rand_arr[:2]
crop_sum = rand_arr[:1] + rand_arr[2:]
print(third_row_sec_col)
print('-----------------------------')
print(crop_first_two_rows)
print('-----------------------------')
print(crop_sum)
print('-----------------------------')
```

#### Task 4.4

```python
rand_arr_sum = np.sum(rand_arr)
rand_arr_avg = np.average(rand_arr)
rand_arr_med = np.median(rand_arr)
rand_arr_std = np.std(rand_arr)
rand_arr_disp = np.var(rand_arr)
print(
    'Array sum elems = ', rand_arr_sum, '\n', 
    'Array avarage value = ', rand_arr_avg, '\n',
    'Array median value = ', rand_arr_med, '\n',
    'Array standart deviation = ', rand_arr_std, '\n',
    'Array variancy = ', rand_arr_disp,'\n',
)
rand_arr_norm = (rand_arr - np.min(rand_arr))/(np.max(rand_arr - np.min(rand_arr)))
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.distplot(
    rand_arr_norm, hist=False, 
    kde=True, color = 'darkblue', 
    hist_kws={'edgecolor':'black'},
    kde_kws={'linewidth': 2}
)
```
