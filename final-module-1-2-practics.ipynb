```python
def mnk(x_, y_):
    n = len(x_)
    m = (np.sum((x - np.mean(x_))*(y_ - np.mean(y_)))) / np.sum((x_ - np.mean(x_))**2)
    c = np.mean(y_) - m*np.mean(x_)
    return m, c
```

```python
# Пример данных для подгонки
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])
m, c = mnk(x, y)
print(f"Найденные коэффициенты линейной функции: m = {m}, c = {c}")
```

```python
data = {
    'Name': ['Petr', 'Anna', 'Maria', 'Alexey', 'Nina'],
    'City': ['Moscow', 'Novosibirsk', 'Samara', 'Sochi', 'Irkutsk'],
    'Age': [41, 28, 0, 22, 33.5]
}

df = pd.DataFrame(data)
df

# Удаление строк с пропусками
df.dropna(inplace=True)

# Удаление дубликатов
df.drop_duplicates(inplace=True)

# Замена нулевых значений средними по столбцу без учета строки с нулевым значением
for column in df.columns:
    if df[column].dtype != 'object':
        df[column] = df[column].replace(0, np.nan)
        df[column].fillna(df[column].mean(skipna=True), inplace=True)

# Вывод итогового DataFrame df
df
```
